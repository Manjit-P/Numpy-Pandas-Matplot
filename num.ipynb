```python
#22. Addition, Subtraction and Multiplication of matrices.
a = np.array([c for c in range(10)]).reshape(2,5)
b = np.array([c for c in range(0,20,2)]).reshape(2,5)
print(a)
print(b)
print(f"Sum: {a+b}")
print(f"Difference: {a-b}")
print(f"Product: {a@b.reshape(5,2)}")
```

```python
#24. Eigen values of a matrix.
from numpy import linalg as LA
A = np.array([c for c in range(1, 10)]).reshape(3,3)
LA.eig(A) #returns eigen values and vectors
LA.eigvals(A) #returns eigen values only.
```

```python
#25. Count frequency of unique values in Numpy array.
b = np.random.choice([0,1], size=10)
unique = np.unique(b)
counts = np.bincount(b)
print("Unique values: ", unique)
print("Count:", counts[unique])
```

```python
#26. Determinant and Inverse of a matrix using numpy
b = np.array([2,1,5,3],).reshape(2,2)
inverse = LA.inv(b)
#inverse.astype(int)
determinant = LA.det(b)
#determinant.astype(int)
```

```python
#27. Multiply matrices of complex numbers using Numpy.
a = np.array([[1+2j, 3-1j], [2+0j, 4+5j]])
b = np.array([[2-3j, 1+4j], [5+2j, -1+1j]])
a @ b
```

```python
#28. Calculate inner, outer and cross products of matrices and vector in Numpy.
A = np.array([c for c in range(6)])
B = np.array([c for c in range(5,11)])
np.inner(A, B)
np.outer(A, B)
np.cross(A.reshape(2,1,3),B.reshape(2,1,3))
```

```python
#29. Convert Covariance matrix to correlation matrix
height = [c for c in range(160, 176, 5)]
weight = [c for c in range(55, 71, 5)]
age = [c for c in range(20, 24)]

# Covariance Matrix
data = np.array([height, weight, age])
cov_matrix = np.cov(data)

# Correlation Matrix
correlation_matrix = np.corrcoef(cov_matrix)
```

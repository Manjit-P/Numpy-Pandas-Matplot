```python
#24. Eigen values of a matrix.
from numpy import linalg as LA
A = np.array([c for c in range(1, 10)]).reshape(3,3)
LA.eig(A) #returns eigen values and vectors
LA.eigvals(A) #returns eigen values only.
```

```python
#28. Calculate inner, outer and cross products of matrices and vector in Numpy.
A = np.array([c for c in range(6)])
B = np.array([c for c in range(5,11)])
np.inner(A, B)
np.outer(A, B)
np.cross(A.reshape(2,1,3),B.reshape(2,1,3))
```

```python
#29. Convert Covariance matrix to correlation matrix
height = [c for c in range(160, 176, 5)]
weight = [c for c in range(55, 71, 5)]
age = [c for c in range(20, 24)]

# Covariance Matrix
data = np.array([height, weight, age])
cov_matrix = np.cov(data)

# Correlation Matrix
correlation_matrix = np.corrcoef(cov_matrix)
```

```python
#33. Replace NaN values with average of columns.
arr = np.array([[1.3, 2.5, 3.6, np.nan], [2.6, 3.3, np.nan, 5.5], [2.1, 3.2, 5.4, 6.5]])
colmean = arr.mean(axis= 0)
inds = np.where(np.isnan(arr))
arr[inds] = np.take(colmean, inds[1])
```

```python
import numpy as np
```

```python
#1. Create an empty and full numpy
a = np.empty((2,2))
b = np.full((2,2), 2)
```

```python
#2. Create a Numpy array filled with zeroes and with ones
a = np.zeros((2,2), dtype= int)
a = np.ones((2,2), dtype= int)
```

```python
#3. Check whether a numpy array contains a specified row.
a = np.array([c for c in range(10)],dtype= int).reshape(5,2)
print([2, 3] in a.tolist())
```

```python
#4. Remove rows in Numpy array that contains non-numeric values.
a = np.array([0, 1, 2, 3, 4, np.nan]).reshape(3, 2)
print(a[np.isfinite(a).all(1)])
```

```python
#5. Combining a one and a two-dimensional Numpy array
oneD = np.array([c for c in range(2)])
twoD = np.array([c for c in range(0,20,2)]).reshape(5, 2)
for a, b in np.nditer([oneD, twoD]):
    print(a,":", b)
```

```python
#6. Get all 2D diagonals of a 3D Numpy array.
threeD = np.array([c for c in range(27)]).reshape(3,3,3)
np.diagonal(threeD, axis1=1, axis2=2)
```

```python
#7. Flatten a matrix in python using numpy.
threeD.flatten()
```

```python
#8. Count number of non-zero values in the array.
choices = [0,1, 2, 3]
a = np.random.choice(choices,size=10)
print(np.count_nonzero(a))
```

```python
#9. Trim leading and/or trailing zeros from a 1D array.
np.trim_zeros(a)
```

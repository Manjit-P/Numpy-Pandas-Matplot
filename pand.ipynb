```python
#5. Convert a list of nested dictionary into pandas dataframe.
rows = []
list = [{
        "Student": [{"Exam": 90, "Grade": "a"},
                    {"Exam": 99, "Grade": "b"},
                    {"Exam": 97, "Grade": "c"},
                    ],
        "Name": "Paras Jain"
        },
        {
        "Student": [{"Exam": 89, "Grade": "a"},
                    {"Exam": 80, "Grade": "b"}
                    ],
        "Name": "Chunky Pandey"
        }
        ]
for data in list:
    data_row = data["Student"]
    _ = data['Name']
    for row in data_row:
        row["Name"] = _
        rows.append(row)

df = pd.DataFrame(rows)
df = df.pivot_table(index="Name", columns="Grade", values="Exam").reset_index()
df.columns = ["Name", "Maths", "Physics", "Chemistry"]
```

```python
#9. Mapping external values to dataframe values in pandas.
data = { 'First_name': ['Liam', 'Emma', 'Noah', 'Olivia', 'Ava'],
         'Last_name': ['Smith', 'Brown', 'Davis', 'Wilson', 'Taylor'],
         'Age': [42, 52, 36, 21, 23],
         'City': ['New York', 'Paris', 'Berlin', 'Madrid', 'Rome'] }
df = pd.DataFrame(data)
new_val = {"Liam": "MBA", "Emma":"PhD", "Noah": "LLB", "Olivia": "B.Tech", "Ava": "MD"}
df["Qualification"] = df["First_name"].map(new_val)

replace_name = {'Liam': 'Lucas', 'Noah': 'Nathanial', 'Olivia': 'Oliver'}
df_replaced = df.replace({"First_name": replace_name})

updates = pd.DataFrame({"First_name":"Nathanail"}, index=[0])
df.update(updates)
```

```python
#5. Convert a list of nested dictionary into pandas dataframe.
rows = []
list = [{
        "Student": [{"Exam": 90, "Grade": "a"},
                    {"Exam": 99, "Grade": "b"},
                    {"Exam": 97, "Grade": "c"},
                    ],
        "Name": "Paras Jain"
        },
        {
        "Student": [{"Exam": 89, "Grade": "a"},
                    {"Exam": 80, "Grade": "b"}
                    ],
        "Name": "Chunky Pandey"
        }
        ]
for data in list:
    data_row = data["Student"]
    _ = data['Name']
    for row in data_row:
        row["Name"] = _
        rows.append(row)

df = pd.DataFrame(rows)
df = df.pivot_table(index="Name", columns="Grade", values="Exam").reset_index()
df.columns = ["Name", "Maths", "Physics", "Chemistry"]
```

```python
#16. Count city population categories
# small city < 100000
# medium city [100000, 1000000]
# large city > 1000000

populations = [50000, 200000, 1500000, 800000, 30000, 2500000]

data = {"city_id": [_ for _ in range(1, 7)],
        "population": populations
        }
df = pd.DataFrame(data)

large = df['population'].get(df['population'] > 1000000).count()
medium = df['population'].get((df['population'] <= 1000000 ) & (df['population'] >= 100000)).count()
small = df['population'].get(df['population'] < 100000).count()

categories = {"category": ['Small City', 'Medium City', 'Large City'],
              "cities_count": [small, medium, large]
             }
summary = pd.DataFrame(categories)
summary[['category', 'cities_count']]
```

```python
#17.Daily products and customer.
# For every (date_id, store_name) pair, calculate:
# unique_products: the number of distinct product_id values.
# unique_customers: the number of distinct customer_id values.

date_ids = ['2021-05-10', '2021-05-10', '2021-05-10', '2021-05-10','2021-05-09', '2021-05-09', 
            '2021-05-09', '2021-05-09', '2021-05-09']
store_names = ['walmart', 'walmart', 'walmart', 'target', 'target', 
               'walmart', 'walmart', 'target', 'walmart']
product_id = [101, 102, 103, 104, 105, 106, 107, 108, 101]
customer_id = [1, 2, 1, 2, 3, 4, 5, 2, 1]

data = {'date_id': date_ids, 'store_name': store_names,
        'product_id': product_id, 'customer_id': customer_id
        }
df = pd.DataFrame(data)

summary = df.groupby(['date_id', 'store_name'], sort=False)[['product_id', 'customer_id']].nunique().reset_index()
summary = summary.rename(columns={'product_id': 'unique_products', 'customer_id': 'unique_customers'})
```

```python
#18. Unique projects worked on by each employee.
import random
employee_ids = [1] * 3 + [2] * 4
project_ids = [101] * 2 + [102] * 2 + [103] * 2 + [104]
dept_ids = [1] * 5 + [2] * 2

for _ in [employee_ids, product_id, dept_ids]:
    random.shuffle(_)
data = {'employee_id': employee_ids,
        'project_id': project_ids,
        'dept_id': dept_ids
        }
df = pd.DataFrame(data)
summary = pd.DataFrame(df.groupby(['employee_id'])['project_id'].nunique())
summary.reset_index().rename(columns={'project_id': 'cnt'})
```

```python
#19. Invalid comments
# commment should be alphanumeric with !,' is allowed.
# len(comment) > 20 : invalid comment
pattern = r"^[A-Za-z0-9 !']+$"
comments = ['Short comment', 'This comment is way too long, sorry']

data = {'comment_id': [1, 2], 'comment': comments}
df = pd.DataFrame(data)

valid = (df['comment'].str.fullmatch(pattern)) & (df['comment'].str.len() <= 20)
invalid_comments = df.loc[~valid]
```

```python
#20. Calculate discount for products.
# no discount to odd product_id
# 100% discount to even product_id with product_name starting with 'A'.
product_ids = [_ for _ in range(1,6)]
categories = ['Carrot', 'Avocado', 'Guava', 'Banana', 'Apple']
prices = [500, 800, 400, 1000, 150]
pattern = r"^A"

data = {'product_id': product_ids,
        'category': categories,
        'price': prices
        }
df = pd.DataFrame(data)

df['discount'] = np.where((df["category"].str.match(pattern)) & 
                          (df["product_id"] % 2 == 0), 0, df["price"])
summary = df[['product_id', 'discount']]
```

```python
import pandas as pd
import numpy as np
```

```python
#1. Pandas dataframe with 2D list.
info = [["Josh","Bradley", 33],
        ["Harry","Lewis", 30],
        ["Ethan","Payne", 31]]
df = pd.DataFrame(info, columns=['Fname', 'Lname', 'Age'])
```

```python
#2. Creating dataframe from dict of narray/lists.
info = [["Josh", "Harry", "Ethan"], [33, 30, 31]]
pairs = {"Name": info[0], "Age": info[1]}
df = pd.DataFrame(pairs)
```

```python
#3. Creating pandas dataframe using list of lists.
Fname = ["Josh", "Harry", "Ethan"]
Lname = ["Bradley", "Lewis", "Payne"]
Age = [33, 30, 31]
info = [[a,b,c] for a,b,c in zip(Fname, Lname, Age)]
columns = ["Fname", "Lname", "Age"]
df = pd.DataFrame(info, columns= columns)
```

```python
#4. Creating a Pandas dataframe using list of tuples.
a = [("Josh", "Bradley", 33),
     ("Harry", "Lewis", 30),
     ("Ethan", "Payne", 31)]
df = pd.DataFrame(a, columns=columns)
```

```python
#5. Convert a list of nested dictionary into pandas dataframe.
rows = []
list = [{
        "Student": [{"Exam": 90, "Grade": "a"},
                    {"Exam": 99, "Grade": "b"},
                    {"Exam": 97, "Grade": "c"},
                    ],
        "Name": "Paras Jain"
        },
        {
        "Student": [{"Exam": 89, "Grade": "a"},
                    {"Exam": 80, "Grade": "b"}
                    ],
        "Name": "Chunky Pandey"
        }
        ]
for data in list:
    data_row = data["Student"]
    _ = data['Name']
    for row in data_row:
        row["Name"] = _
        rows.append(row)

df = pd.DataFrame(rows)
df = df.pivot_table(index="Name", columns="Grade", values="Exam").reset_index()
df.columns = ["Name", "Maths", "Physics", "Chemistry"]
```

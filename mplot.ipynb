```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Plot horizontal & vertical line.
plt.hlines(y=[1, 2, 3], xmin=0, xmax=5, colors=['b', 'g', 'r'],
           linestyles =['-', '--', ':'])
plt.vlines(x=[2, 4, 6], ymin=0, ymax=6, colors=['b', 'g', 'r'], 
           linestyles =['-', '--', ':'])
plt.show()
```

```python
#2. Change line opacity.
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y, color='blue', alpha=0.3, label='Alpha= 0.5')
plt.legend()
plt.show()
```

```python
#3. Increase the thickness of a line
plt.plot(x, y, color='yellow', linewidth=10, marker='o', markerfacecolor='red')
plt.show()
```

```python
#4. Fill between multiple lines.
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()
```

```python
#5. Draw horizontal bar chart
x =['a', 'b', 'c', 'd']
y =np.random.randint(0, 100, 4)

plt.barh(x, y)
plt.show()
```

```python
#6. Display the value of each bar in a bar chart.
bars = plt.bar(x, y)
plt.bar_label(bars, label_type='edge')
plt.show()
```

```python
#7. Annotate bars in barplot.
names = ['Alex', 'Bob', 'Clarein', 'Dexter']
marks = [45, 23, 78, 65]

datas = {'name': names, 'mark': marks}
df = pd.DataFrame(datas)

bars = plt.bar(df['name'], df['mark'])
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks obtained by students')
#plt.bar_label(bars)
for p in bars:
    plt.annotate(format(p.get_height())+" stars", (p.get_x() + p.get_width()/2, p.get_height()))
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Plot horizontal & vertical line.
plt.hlines(y=[1, 2, 3], xmin=0, xmax=5, colors=['b', 'g', 'r'],
           linestyles =['-', '--', ':'])
plt.vlines(x=[2, 4, 6], ymin=0, ymax=6, colors=['b', 'g', 'r'], 
           linestyles =['-', '--', ':'])
plt.show()
```

```python
#2. Change line opacity.
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y, color='blue', alpha=0.3, label='Alpha= 0.5')
plt.legend()
plt.show()
```

```python
#3. Increase the thickness of a line
plt.plot(x, y, color='yellow', linewidth=10, marker='o', markerfacecolor='red')
plt.show()
```

```python
#4. Fill between multiple lines.
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()
```

```python
#5. Draw horizontal bar chart
x =['a', 'b', 'c', 'd']
y =np.random.randint(0, 100, 4)

plt.barh(x, y)
plt.show()
```

```python
#6. Display the value of each bar in a bar chart.
bars = plt.bar(x, y)
plt.bar_label(bars, label_type='edge')
plt.show()
```

```python
#7. Annotate bars in barplot.
names = ['Alex', 'Bob', 'Clarein', 'Dexter']
marks = [45, 23, 78, 65]

datas = {'name': names, 'mark': marks}
df = pd.DataFrame(datas)

bars = plt.bar(df['name'], df['mark'])
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks obtained by students')
#plt.bar_label(bars)
for p in bars:
    plt.annotate(format(p.get_height())+" stars", (p.get_x() + p.get_width()/2, p.get_height()))
plt.show()
```

```python
#8. Create a scatterplot.
x = np.arange(1, 10)
y1 = x ** 2
y2 = x ** 3

plt.scatter(x, y1, label='square of x')
plt.scatter(x, y2, label='cube of x')

plt.legend()
plt.show()
```

```python
#9. Create scatterplot with several colors.
x = np.arange(1,6)
y = np.random.randint(0, 20, 5)

plt.scatter(x, y, c='green')

x = np.random.randint(10, 20, 5)
y = np.random.randint(10, 20, 5)

plt.scatter(x, y, c='red')

plt.show()
```

```python
#10. Increase the size of scatter ponts.
x = np.random.randint(1, 20, 10)
y = np.random.randint(1, 20, 10)

plt.scatter(x, y, s=300, c='r')
plt.show()
```

```python
#11. Connect scatterplot point with line.
x = np.arange(-25, 25)
y = x ** 2

plt.scatter(x, y, c='b')
plt.plot(x,y)
plt.show()
```

```python
# 12. Change the x or y interval of a figure.

x = np.arange(1, 25)
y = x + 2

plt.xlim([0, 40])
plt.ylim([0, 100])
plt.plot(x, y)

plt.show()
```

```python
#13. Reverse axes.
x = np.linspace(0, 10, 100)
y1 = x ** 2
y2 = np.exp(x)
fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(x, y1, label='x^2', c='b')
ax1.plot(x, y2, label='exp(x)', c='r')
ax1.set_title('Normal')
ax1.legend()
ax2.plot(x, y1, label='x^2', c='b')
ax2.plot(x, y2, label='exp(x)', c='r')
ax2.set_title('Reversed')
ax2.legend()
ax2.invert_xaxis()
ax2.invert_yaxis()
plt.show()
```

```python
#14. Use different y-axes on the left and right of same graph.
x = np.linspace(0, 10, 100)
y1 = x ** 2
y2 = np.exp(x)
fig, ax1 = plt.subplots()
ax1.plot(x, y1, label='x^2', c='b')
ax1.plot(x, y2, label='exp(x)', c='r')
ax1.set_title('Normal')
ax1.legend()
ax2 = ax1.twinx()

ax2.plot(x, -y2, label='-exp(x)', c='r')
ax2.set_title('Reversed')
ax2.legend()

plt.show()
```
